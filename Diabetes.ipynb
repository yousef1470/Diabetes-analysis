# Drop rows equal 0

```python
data['Insulin'].replace(0, pd.NA, inplace=True)
data['SkinThickness'].replace(0, pd.NA, inplace=True)
data['Pregnancies'].replace(0, pd.NA, inplace=True)
data['Glucose'].replace(0, pd.NA, inplace=True)
data = data.dropna()
```

# Glucose Categories

```python
# Over 126 mg/dL is considered diabetic, 100-125 mg/dL is considered prediabetic, and below 100 mg/dL is considered normal
bins = [0, 100, 126, 200]
labels = ['Normal', 'Prediabetes', 'Diabetes']
data['Glucose_Category'] = pd.cut(data['Glucose'], bins=bins, labels=labels)
glucose_counts = data['Glucose_Category'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
sns.barplot(x=glucose_counts.index, y=glucose_counts.values, palette='coolwarm')
plt.title('Count of Individuals in Glucose Categories')
plt.xlabel('Glucose Category')
plt.ylabel('Count')
plt.show()
```

# Percentage of individuals with high glucose levels

```python
high_glucose = data['Glucose'] > 126
glucose_counts = high_glucose.value_counts()
plt.figure(figsize=(6, 6))
plt.pie(glucose_counts, labels=['Glucose ≤ 126', 'Glucose > 126'], autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Percentage of Individuals with High Glucose Levels')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(glucose_counts, labels=glucose_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'orange', 'lightcoral'])
plt.title('Proportion of Individuals in Glucose Categories')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Glucose_Category', hue='Outcome', data=data, palette='Set2')
plt.title('Count of Individuals in Glucose Categories by Diabetes Status')
plt.xlabel('Glucose Category')
plt.ylabel('Count')
plt.legend(title='Outcome', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

# BMI categories

```python
bins = [0, 18.5, 25, 30, 100]
labels = ['Underweight', 'Normal weight', 'Overweight', 'Obese']
data['BMI_Category'] = pd.cut(data['BMI'], bins=bins, labels=labels)

# Count individuals in each BMI category
bmi_counts = data['BMI_Category'].value_counts().sort_index()

# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=bmi_counts.index, y=bmi_counts.values, palette='Set2')
plt.title('Count of Individuals in BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'orange', 'lightcoral'])
plt.title('Proportion of Individuals in BMI Categories')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='BMI_Category', hue='Outcome', data=data, palette='Set2')
plt.title('Count of Individuals in BMI Categories by Diabetes Status')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.legend(title='Outcome', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

# Blood Pressure Categories

```python
# Define Blood Pressure categories
bins = [0, 80, 90, 100, 200]
labels = ['Normal', 'Elevated', 'Hypertension Stage 1', 'Hypertension Stage 2']
data['BP_Category'] = pd.cut(data['BloodPressure'], bins=bins, labels=labels)

bp_counts = data['BP_Category'].value_counts().sort_index()

# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=bp_counts.index, y=bp_counts.values, palette='Set2')
plt.title('Count of Individuals in Blood Pressure Categories')
plt.xlabel('Blood Pressure Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(bp_counts, labels=bp_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'yellow', 'orange', 'lightcoral'])
plt.title('Proportion of Individuals in Blood Pressure Categories')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='BP_Category', hue='Outcome', data=data, palette='Set2')
plt.title('Count of Individuals in Blood Pressure Categories by Diabetes Status')
plt.xlabel('Blood Pressure Category')
plt.ylabel('Count')
plt.legend(title='Outcome', labels=['No Diabetes', 'Diabetes'])
plt.show()
```
